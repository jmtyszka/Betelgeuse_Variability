```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
%matplotlib inline
```

```python
# Load data
fname = "AAVSO_Betelgeuse_50yrs.csv"

df = pd.read_csv(fname)

# Include only broadband visible observations
df = df[df["Band"] == "Vis."]

# Cast magnitude to numeric where possible
df["Magnitude"] = pd.to_numeric(df["Magnitude"], errors="coerce")

# Convert Julian dates to YYYY-MM-DD
df["Date"] = dt. df["JD"]
```
